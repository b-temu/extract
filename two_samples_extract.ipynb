```python
import json

# Initialize variables
query_cluster = "c00015_NODE_63"
genes = []
significant_hits = []

# Read data from file
with open(file_path_txt, 'r') as f:
    lines = f.readlines()
    
    # Find the starting line of the table
    start_index_table = 0
    for idx, line in enumerate(lines):
        if "Table of genes, locations, strands and annotations of query cluster:" in line:
            start_index_table = idx + 1  # Skip the header and move to the first gene line
            break
    
    # Process each line from the table
    for line in lines[start_index_table:]:
        # Strip any leading/trailing whitespace and split by whitespace
        parts = line.strip().split()
        
        # Ensure the line is in the expected format
        if len(parts) >= 4:
            gene_name = parts[0]
            start = int(parts[1])
            end = int(parts[2])
            strand = parts[3]
            
            # Construct gene dictionary
            gene_dict = {
                "gene": gene_name,
                "start": start,
                "end": end,
                "strand": strand
            }
            
            # Append gene dictionary to genes list
            genes.append(gene_dict)
        else:
            # Stop processing when encountering the significant hits section
            break
    
    # Find the starting line of the significant hits
    start_index_hits = 0
    for idx, line in enumerate(lines):
        if "Significant hits:" in line:
            start_index_hits = idx + 1  # Skip the header and move to the first hit line
            break
    
    # Process the first 5 significant hits
    for line in lines[start_index_hits:start_index_hits + 5]:
        hit = line.strip()
        hit_name = hit.split("\t")[0][2:]
        hit_desc = hit.split("\t")[1]
        significant_hits.append({
            "hit_name": hit_name,
            "hit_desc": hit_desc

        })

# Display extracted genes (for verification)
print("Genes:")
for gene in genes:
    print(gene)

# Display first 5 significant hits (for verification)
print("\nFirst 5 Significant Hits:")
for hit in significant_hits:
    print(hit)

# Optionally, convert to JSON and write to file
json_data = {
    # "query_cluster": query_cluster,
    "genes": genes,
    "significant_hits": significant_hits[:5]  # Only include the first 5 hits in JSON
}

# Convert Python dictionary to JSON formatted string
json_str = json.dumps(json_data, indent=2)

# Write JSON to output file
with open('output.json', 'w') as outfile:
    outfile.write(json_str)

print("\nJSON structure written to output.json")
```

```python
import json

def text_to_json(input_file, output_file, method='lines'):
    with open(input_file, 'r') as file:
        if method == 'lines':
            # Each line becomes an element in a JSON array
            data = file.readlines()
            json_data = json.dumps(data)
        elif method == 'whole':
            # The entire file becomes a single JSON string
            data = file.read()
            json_data = json.dumps({"content": data})
        elif method == 'key_value':
            # Assumes each line is in "key: value" format
            data = {}
            for line in file:
                key, value = line.strip().split(':', 1)
                data[key.strip()] = value.strip()
            json_data = json.dumps(data)
    
    with open(output_file, 'w') as file:
        file.write(json_data)

# Usage examples:

file_path_json = "./output.json"
text_to_json(file_path_txt, file_path_json, 'whole')


# text_to_json('input.txt', 'output.json', 'whole')
# text_to_json('input.txt', 'output.json', 'key_value')
```
